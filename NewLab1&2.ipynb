```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('/content/DATA (1).csv')
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Names:")
for col in df.columns:
    print(col)
```

```python
df.info()
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
df.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9])
```

```python
df_clean = df.drop(columns=['STUDENT ID'])
df_clean.head()
```

```python
plt.figure(figsize=(6,4))
df['GRADE'].plot.density()
plt.title('Density Plot of Grade')
plt.xlabel('Grade')
plt.show()
```

```python
sample_cols = df_clean.columns[:6]  # first 6 features
sns.pairplot(df_clean[sample_cols])
```

```python
correlation_matrix = df_clean.corr()
correlation_matrix
```

```python
fig, axs = plt.subplots(len(df_clean.columns), 1, figsize=(7, 20), dpi=90)

for i, col in enumerate(df_clean.columns):
    axs[i].boxplot(df_clean[col], vert=False)
    axs[i].set_ylabel(col)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df_clean['GRADE'], bins=5)
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Histogram of Grade Distribution')
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_clean.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
X = df_clean.drop(columns=['GRADE'])
y = df_clean['GRADE']

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
corr_target = df_clean.corr()['GRADE'].sort_values(ascending=False)
corr_target
```
